```python
# import libraries -- this is like installing tools
import pandas as pd        # pandas = works with data tables like Excel
import matplotlib.pyplot as plt  # matplotlib = creates charts
import seaborn as sns      # seaborn = creates beautiful charts

print("Libraries imported!")  # print = shows a message on screen
```

```python
# ================================================
# Step 2 - Load the dataset
# ================================================

# pd.read_csv = reads the CSV file and stores it in 'df'
# df = dataframe = think of it like an Excel table in Python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# .shape = shows how many rows and columns we have
print("Dataset Shape:", df.shape)

# .head() = shows first 5 rows of data -- like SELECT * LIMIT 5 in SQL
df.head()
```

```python
# ================================================
# Step 3 - Explore the dataset
# ================================================

# .info() = shows column names, data types and missing values
# like DESCRIBE table in SQL
print("=== Dataset Info ===")
df.info()

print("\n=== Missing Values ===")
# .isnull().sum() = counts missing values in each column
print(df.isnull().sum())

print("\n=== Basic Statistics ===")
# .describe() = shows min, max, average of all number columns
df.describe()
```

```python
# ================================================
# Q1 - What is the overall attrition rate?
# Business Question: How many employees left the company?
# ================================================

# count how many Yes and No in Attrition column
# value_counts() = like GROUP BY in SQL
attrition_count = df['Attrition'].value_counts()
print("=== Attrition Count ===")
print(attrition_count)

# calculate attrition rate percentage
# len(df) = total rows = like COUNT(*) in SQL
attrition_rate = (df['Attrition'].value_counts()['Yes'] / len(df)) * 100
print(f"\nAttrition Rate: {attrition_rate:.2f}%")

# create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(attrition_count,
        labels=['No', 'Yes'],
        autopct='%1.1f%%',
        colors=['#2196F3', '#FF5722'])
plt.title('Overall Employee Attrition Rate')
plt.show()
```

```python
# ================================================
# Q2 - Which department has highest attrition?
# Business Question: Where is the company losing most people?
# ================================================

# filter only employees who left (Attrition = Yes)
# this is like WHERE Attrition = 'Yes' in SQL
attrition_dept = df[df['Attrition'] == 'Yes']['Department'].value_counts()
print("=== Attrition by Department ===")
print(attrition_dept)

# create bar chart
plt.figure(figsize=(8, 5))
sns.countplot(data=df,           # dataset
              x='Department',    # x axis
              hue='Attrition',   # split by Yes/No
              palette=['#2196F3', '#FF5722'])  # colors
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()
```

```python
# ================================================
# Q3 - Does age affect attrition?
# Business Question: Which age group leaves the most?
# ================================================

plt.figure(figsize=(10, 5))
sns.histplot(data=df,          # dataset
             x='Age',          # x axis = age
             hue='Attrition',  # split by Yes/No
             bins=20,          # number of bars
             palette=['#2196F3', '#FF5722'])
plt.title('Attrition by Age Group')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.show()

# average age of employees who left vs stayed
avg_age = df.groupby('Attrition')['Age'].mean()
print("\n=== Average Age ===")
print(avg_age)
```

```python
# ================================================
# Q4 - Does salary affect attrition?
# Business Question: Are low salary employees leaving more?
# ================================================

# average salary of who left vs stayed
avg_salary = df.groupby('Attrition')['MonthlyIncome'].mean()
print("=== Average Monthly Income ===")
print(avg_salary)

# box plot to compare salary distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=df,
            x='Attrition',
            y='MonthlyIncome',
            palette=['#FF5722', '#2196F3'])
plt.title('Monthly Income vs Attrition')
plt.xlabel('Attrition (Yes = Left, No = Stayed)')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# ================================================
# Q5 - Does overtime affect attrition?
# Business Question: Do overworked employees leave more?
# ================================================

# count attrition by overtime
overtime_attrition = df.groupby(['OverTime', 'Attrition']).size().unstack()
print("=== Overtime vs Attrition ===")
print(overtime_attrition)

# calculate overtime attrition rate
overtime_rate = df[df['OverTime'] == 'Yes']['Attrition'].value_counts(normalize=True) * 100
print(f"\nAttrition rate WITH overtime: {overtime_rate['Yes']:.2f}%")

no_overtime_rate = df[df['OverTime'] == 'No']['Attrition'].value_counts(normalize=True) * 100
print(f"Attrition rate WITHOUT overtime: {no_overtime_rate['Yes']:.2f}%")

# bar chart
plt.figure(figsize=(7, 5))
sns.countplot(data=df,
              x='OverTime',
              hue='Attrition',
              palette=['#2196F3', '#FF5722'])
plt.title('Overtime vs Attrition')
plt.xlabel('Overtime')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()
```

```python
# ================================================
# Q6 - Which job role has highest attrition?
# Business Question: Which roles are hardest to retain?
# ================================================

# attrition rate by job role
job_attrition = df.groupby('JobRole')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).round(2).sort_values(ascending=False)

print("=== Attrition Rate by Job Role ===")
print(job_attrition)

# bar chart
plt.figure(figsize=(12, 6))
job_attrition.plot(kind='bar',
                   color='#FF5722',
                   edgecolor='white')
plt.title('Attrition Rate by Job Role (%)')
plt.xlabel('Job Role')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# ================================================
# Q7 - Does work life balance affect attrition?
# Business Question: Do unhappy employees leave more?
# ================================================

# work life balance is rated 1-4
# 1=Bad, 2=Good, 3=Better, 4=Best
wlb_attrition = df.groupby('WorkLifeBalance')['Attrition'].apply(
    lambda x: (x == 'Yes').sum() / len(x) * 100
).round(2)

print("=== Attrition Rate by Work Life Balance ===")
print("(1=Bad, 2=Good, 3=Better, 4=Best)")
print(wlb_attrition)

# bar chart
plt.figure(figsize=(8, 5))
sns.countplot(data=df,
              x='WorkLifeBalance',
              hue='Attrition',
              palette=['#2196F3', '#FF5722'])
plt.title('Work Life Balance vs Attrition')
plt.xlabel('Work Life Balance (1=Bad, 2=Good, 3=Better, 4=Best)')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()
```

```python
# ================================================
# Q8 - How does years at company affect attrition?
# Business Question: When are employees most likely to leave?
# ================================================

# average years at company
avg_years = df.groupby('Attrition')['YearsAtCompany'].mean()
print("=== Average Years at Company ===")
print(avg_years)

# histogram
plt.figure(figsize=(10, 5))
sns.histplot(data=df,
             x='YearsAtCompany',
             hue='Attrition',
             bins=20,
             palette=['#2196F3','#FF5722'])
plt.title('Years at Company vs Attrition')
plt.xlabel('Years at Company')
plt.ylabel('Number of Employees')
plt.show()
```
